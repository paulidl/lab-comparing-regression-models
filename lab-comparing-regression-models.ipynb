#### Taking the out outliers.

```python
def remove_outliers(data, col):
    Q3 = data[col].quantile(0.75)
    Q1 = data[col].quantile(0.25) 
    iqr = Q3 - Q1
    upper_limit = Q3 + 1.5 * iqr
    lower_limit = Q1 - 1.5 * iqr
    data_final=data[~((data[col] < lower_limit)|(data[col] > upper_limit))]
    data_final.reset_index(drop=True, inplace=True) 
    return data_final

customer_df= remove_outliers(customer_df, "total_claim_amount")
customer_df.head()
```

```python
def diff_function(data):
        df_categoricals = pd.DataFrame()
        df_continuous = pd.DataFrame()
        df_discretes = pd.DataFrame()
        for i in data:  
            if np.dtype(data[i]) == 'object':
                df_categoricals[i] = data[i]
            elif len(data[i].unique()) > 20 or np.dtype(data[i]) == 'float':  #(len(data[i].unique())*0.8):
                df_continuous[i] = data[i]
            else: 
                df_discretes[i] = data[i]
        return df_categoricals, df_continuous, df_discretes
    
df_categoricals, df_continuous, df_discretes = diff_function(df)
```

#### 3. Great! Now define a function that takes a list of models and train (and tests) them so we can try a lot of them without repeating code.

```python
def modeling(y, X, models=[], test_size=0.3):
    for model in models:
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
        model.fit(X_train, y_train )
        y_pred = model.predict(X_test)
        r2 = r2_score(y_test, y_pred)
        MSE = mean_squared_error(y_test, y_pred, squared=True)
        RMSE = mean_squared_error(y_test, y_pred, squared=False)
        MAE = mean_absolute_error(y_test, y_pred)
        print(model, 'metrics are: '), print("R2 score =", round(r2,2)), print("RMSE =", round(RMSE,2)), print("MSE =", round(MSE,2)), print("MAE =", round(MAE,2))
    return y_pred, y_test, r2
```

#### 4. Use the function to check LinearRegressor and KNeighborsRegressor.

```python
predictions, y_test, r2 = modeling(y, X, models=[LinearRegression(), KNeighborsRegressor(n_neighbors=8)], test_size=0.3)
```

#### 5. You can check also the MLPRegressor for this task!

```python
predictions, y_test, r2 = modeling(y, X, models=[MLPRegressor(max_iter=100)], test_size=0.3)
```
